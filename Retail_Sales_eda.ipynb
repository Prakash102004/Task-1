```python
import pandas as pd

# Load the dataset
df = pd.read_csv('menu.csv')

# Display the first 5 rows
print(df.head())

# Basic info and null check
print(df.info())
print(df.isnull().sum())
```

```python
# Summary statistics for numerical columns
print(df.describe())

# Summary statistics for all columns (including object type)
print(df.describe(include='all'))
```

```python
 Suppose there's a column 'Date' and 'Sales'
df['Date'] = pd.to_datetime(df['Date'])  # replace 'Date' with actual column name
df.set_index('Date', inplace=True)

# Resample to monthly sales
monthly_sales = df['Sales'].resample('M').sum()

# Plot the trend
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
# Top-selling menu items (by frequency of appearance in dataset)
top_items = df['Item'].value_counts().head(10)
print("Top Menu Items:")
print(top_items)

# Top categories (most common food types)
top_categories = df['Category'].value_counts().head(10)
print("\nTop Menu Categories:")
print(top_categories)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate top 10 menu items (by frequency)
top_items = df['Item'].value_counts().head(10)

# Bar chart: Top 10 Menu Items
plt.figure(figsize=(10, 5))
top_items.plot(kind='bar', color='skyblue')
plt.title('Top 10 Menu Items')
plt.xlabel('Item')
plt.ylabel('Number of Appearances')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Heatmap: Correlation between numeric features
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap of Nutritional Info')
plt.tight_layout()
plt.show()

```
